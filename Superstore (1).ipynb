```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Sample - Superstore.csv', encoding='latin1')
```

```python
df.info()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
```

```python
df['Ship Date']=pd.to_datetime(df['Ship Date'],errors='coerce')
```

```python
df.info()
```

```python
df.drop(['Order ID'],axis=1,inplace=True)
```

```python
df.drop(['Row ID'],axis=1,inplace=True)
```

```python
df.drop(['Customer ID'],axis=1,inplace=True)
```

```python
df.drop(['Postal Code'],axis=1,inplace=True)
```

```python
df.drop(['Product ID'],axis=1,inplace=True)
```

```python
df
```

```python
df['Country'].unique()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df[['Sales', 'Profit']].sum()
```

```python
sales = 2296919
profit = 286409

profit_percentage = (profit / sales) * 100
print(f"Profit Percentage: {profit_percentage:.2f}%")
```

```python
import matplotlib.pyplot as plt

sales = 2296919
profit = 286409

# القيم
labels = ['Sales', 'Profit']
values = [sales, profit]
colors = ['skyblue', 'lightgreen']

plt.figure(figsize=(6, 5))
plt.bar(labels, values, color=colors, edgecolor='black')

plt.title('Total Sales vs Profit', fontsize=14)
plt.ylabel('Amount')
plt.grid(axis='y', linestyle='--', alpha=0.7)

for i, v in enumerate(values):
    plt.text(i, v + 50000, f"{v:,}", ha='center', fontweight='bold')

plt.tight_layout()
plt.show()

```

```python
df.groupby('Region')['Profit'].sum().sort_values(ascending=False)
```

```python
df.groupby('Category')['Sales'].sum().sort_values(ascending=False)[0:5]

```

```python
df.groupby('Sub-Category')['Quantity'].sum().sort_values(ascending=False)
```

```python



discount_by_segment = df.groupby('Segment')['Discount'].mean()


plt.figure(figsize=(8, 5))
discount_by_segment.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title('Average Discount by Segment', fontsize=14)
plt.xlabel('Segment')
plt.ylabel('Average Discount')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# what is The day with the highest sales?
```

```python
sales_by_date = df.groupby('Order Date')['Sales'].sum()

top_sales_day = sales_by_date.idxmax()
top_sales_amount = sales_by_date.max()

print(f"The day with the highest sales was: {top_sales_day.date()} with total sales of ${top_sales_amount:.2f}")
```

```python
# Which product category generates the most sales?
```

```python

category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
```

```python

category_sales.plot(kind='bar', color='mediumseagreen', edgecolor='black', figsize=(8,5))
plt.title('Total Sales by Category', fontsize=14)
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# What is the trend of sales over time?
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum().plot(kind='line', figsize=(12,5))
```

```python
# How does average profit vary by state?
```

```python
df.groupby('State')['Profit'].mean().sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt

state_profit = df.groupby('State')['Profit'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
state_profit.plot(kind='bar', color='mediumseagreen', edgecolor='black')

plt.title('Average Profit by State', fontsize=14)
plt.xlabel('State')
plt.ylabel('Average Profit')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
